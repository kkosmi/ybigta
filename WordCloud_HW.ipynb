```
mask_array = np.array(Image.open("/content/gdrive/MyDrive/pngegg.png"))
```

```
stopwords.update(["Mr","said","Mrs"])
pride_wd = WordCloud(background_color = 'white',
                     collocations = False,
                     stopwords = stopwords,
                     mask = mask_array,
                     colormap = 'tab10'
                    )
pride_wd.generate(pride_and_prejudice)
pride_wd.to_image()
```
