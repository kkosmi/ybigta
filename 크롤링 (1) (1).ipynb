## requests 모듈
=HTTP 통신을 대신 수행해주는 모듈

get = 데이터를 URL에 포함하여 전달(대부분의 요청에서 사용 됨)  
post = 데이터를 Form data에 포함하여 전달(민감한 정보(로그인)에서 사용 됨)

### get 요청하기

```python
url = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query=%EC%A0%9C%EC%A3%BC%EB%8F%84&oquery=%EC%A0%9C%EC%A3%BC%EB%8F%84+%EC%97%AC%ED%96%89&tqi=htg9ydp0YiRsseJsM%2FNsssssshV-011537'
resp = requests.get(url)

print(resp) #응답 코드
resp.text #HTML 코드
```

#### HTTP 상태 코드
 - 1xx (정보): 요청을 받았으며 프로세스를 계속한다
 - 2xx (성공): 요청을 성공적으로 받았으며 인식했고 수용하였다
 - 3xx (리다이렉션): 요청 완료를 위해 추가 작업 조치가 필요하다
 - 4xx (클라이언트 오류): 요청의 문법이 잘못되었거나 요청을 처리할 수 없다
 - 5xx (서버 오류): 서버가 명백히 유효한 요청에 대해 충족을 실패했다

[출처: 위키피디아](https://ko.wikipedia.org/wiki/HTTP_%EC%83%81%ED%83%9C_%EC%BD%94%EB%93%9C)

## BeautifulSoup 모듈
=Requests로 가져온 HTML을 분석하기 쉽게 바꿔주는 모듈


*CSS, 속성 값, 정규표현식 등 다양한 방법으로 값을 찾을 수 있다.

## 기본 함수

```python
#from bs4 import BeautifulSoup
```

```python
html = '''
<html>
  <head>
    <title>Example HTML</title>
  </head>
  <body>
    <div id='first' class='Example'>
      <h3 title='Good Content Title'>Contents Title</h3>
      <p>Test contents</p>
    </div>
    <div id='second' class='Example'>
      <p>Text1</p>
      <p>Text2</p>
      <p>Text3</p>
    </div>
  </body>
</html>'''
```

```python
soup = BeautifulSoup(html, 'html.parser') 
#soup = BeautifulSoup(resp.text, 'html.parser')
```

## find, find_all 함수
= HTML tag와 속성을 활용해 검색

## get_text 함수
= Tag안에 있는 value 찾기 (부모 tag의 경우 자식 tag의 value를 모두 반환)

```python
tag = soup.find('h3')
print(tag)
tag.get_text()
```

```python
tag = soup.find('p')
print(tag)
tag.get_text()
```

```python
soup.find_all('p')
#find_all은 list 형태로 값 반환
#list 형태이므로 get_text()바로 쓰지 못한다.
```

```python
soup.find_all('p')[2].get_text()
```

```python
soup.find_all('p', limit =2)
```

```python
tag = soup.find('div')
print(tag)
tag.get_text()
```

```python
tag = soup.find('div', id='second')
print(tag)
tag.get_text()
```

```python
attrs = {
    'id' : 'second',
    'class' : 'Example'
}
soup.find('div', attrs = attrs)
```

```python
driver.get('https://www.naver.com/')  #해당링크로 이동
```

```python
#검색창에 커서 두기
search = driver.find_element_by_id('query')
time.sleep(1)
```

```python
#검색어 입력
search.send_keys('제주도')
```

```python
time.sleep(1)
#Keys.RETURN = Enter 입력
search.send_keys(Keys.RETURN)
time.sleep(1)
```

```python
#View목록
search = driver.find_element_by_link_text('뉴스')
search.click()
```

```python
driver.get('https://finance.naver.com/')

all_title=[]

for i in range(1,5):
    i=str(i)
    css = '#content > div.article > div.section > div.news_area > div > ul > li:nth-child(' + i + ') > span > a'
    content = driver.find_element_by_css_selector(css)
    all_title.append(content.text)

print(all_title)
```
